Load

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/cleaned/cleaned_sales.csv')
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
print(df.shape)
```

Monthly Revenue Trend

```python
monthly = df.groupby('YearMonth')['Revenue'].sum().reset_index()

plt.figure(figsize=(14,5))
plt.plot(monthly['YearMonth'], monthly['Revenue'], marker='o', color='steelblue')
plt.xticks(rotation=45)
plt.title('Monthly Revenue Trend')
plt.tight_layout()
plt.savefig('../screenshots/monthly_revenue.png')
plt.show()
```

Top 10 Products

```python
top_products = df.groupby('Description')['Revenue'].sum().nlargest(10).reset_index()

plt.figure(figsize=(10,6))
sns.barplot(data=top_products, x='Revenue', y='Description', palette='Blues_r')
plt.title('Top 10 Products by Revenue')
plt.tight_layout()
plt.savefig('../screenshots/top_products.png')
plt.show()
```

Top 10 Countries

```python
top_countries = df.groupby('Country')['Revenue'].sum().nlargest(10).reset_index()

plt.figure(figsize=(10,6))
sns.barplot(data=top_countries, x='Revenue', y='Country', palette='Greens_r')
plt.title('Top 10 Countries by Revenue')
plt.tight_layout()
plt.savefig('../screenshots/top_countries.png')
plt.show()
```

Revenue By Day

```python
day_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
day_revenue = df.groupby('DayName')['Revenue'].sum().reindex(day_order).reset_index()

plt.figure(figsize=(10,5))
sns.barplot(data=day_revenue, x='DayName', y='Revenue', palette='Oranges_r')
plt.title('Revenue by Day of Week')
plt.tight_layout()
plt.savefig('../screenshots/day_revenue.png')
plt.show()
```

Exporting Monthly for Power BI

```python
monthly.to_csv('../exports/monthly_revenue.csv', index=False)
top_products.to_csv('../exports/top_products.csv', index=False)
top_countries.to_csv('../exports/top_countries.csv', index=False)

print("✅ EDA exports done!")
```
